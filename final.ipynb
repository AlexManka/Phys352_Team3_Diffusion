```python
import numpy as np 
import matplotlib.pyplot as plt
import random
```

```python
# Importing data for 100 walkers taking 1000 random steps each, initialized at origin

x, y, step = np.loadtxt("data7/7_5.dat", unpack=True)
```

```python
# Creates a plot of the first 20 steps for all of the walkers. 

MAX = 1000 # number of steps
walkers = round(len(x)/MAX) # calculates number of walkers
colors = [] # list for storing randomly generated colors

plt.figure(figsize=(12.8, 12.8))
for i in range(walkers):
    # generates a random color using hexadecimal 
    colors.append("#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])) 
    
    # plots the first 20 steps of each walker with a random color
    plt.plot(x[i*MAX: i*MAX+15], y[i*MAX: i*MAX+15], lw=1, color=colors[i])
```

```python
# Creates an animation of the first 20 steps of all walkers (initiated at origin)

from celluloid import Camera

fig = plt.figure()
camera = Camera(fig)
for i in range(200):
    for j in range(walkers):
        plt.plot(x[j*MAX+i], y[j*MAX+i], "o", color=colors[j])
        plt.gca().set_aspect(1.0)
    camera.snap()

animation = camera.animate(interval=50)
animation.save("randomwalk_origin.gif")
    
```

```python
# Importing data for 500 walkers taking 1000 random steps each, with a Gaussian distribution of initial positions.

x1, y1, step1 = np.loadtxt("data7/gauss.dat", unpack=True)
```

```python
# Creates a plot of the initial positions of all walkers, which have a Gaussian distribution

MAX = 1000 # number of steps
molecules = round(len(x1)/MAX) # calculates number of walkers

plt.gca().set_aspect(1.0)
plt.xlim(-20, 20)
plt.ylim(-20, 20)
for i in range(molecules):
    plt.plot(x1[i*MAX], y1[i*MAX], ".", color="orange")
```

```python
# Creates an animation of the first 200 steps of all walkers (initiated with Gaussian distribution)

from celluloid import Camera

fig = plt.figure()
camera = Camera(fig)
for i in range(200):
    for j in range(molecules):
        plt.plot(x1[j*MAX+i], y1[j*MAX+i], ".", color="orange")
        plt.gca().set_aspect(1.0)
    camera.snap()

animation = camera.animate(interval=50)
animation.save("randomwalk_gauss.gif")
    
```
